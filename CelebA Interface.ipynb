```python
import pandas as pd
import numpy as np
import random
import os
import csv
from tkinter import *   
from PIL import ImageTk, Image


#Reading CelebA dataset
data = pd.read_csv('C:\\Users\lenovo\\Desktop\\list_attr_celeba.csv')

global tempDataset
tempDataset = {}
dataset = pd.DataFrame(tempDataset)

#Assigning columns for emotional attributes
excitedArray = np.zeros(len(data))
delightedArray = np.zeros(len(data))
happyArray = np.zeros(len(data))
contentArray = np.zeros(len(data))
relaxedArray = np.zeros(len(data))
calmArray = np.zeros(len(data))
tiredArray = np.zeros(len(data))
boredArray = np.zeros(len(data))
depressedArray = np.zeros(len(data))
frustratedArray = np.zeros(len(data))
angryArray = np.zeros(len(data))
tenseArray = np.zeros(len(data))

dataArousal = data.assign(Excited = excitedArray, Delighted = delightedArray, Happy = happyArray,
                          Content = contentArray, Relaxed = relaxedArray, Calm = calmArray,
                          Tired = tiredArray, Bored = boredArray, Depressed = depressedArray,
                          Frustrated = frustratedArray, Angry = angryArray, Tense = tenseArray)

#Generate random image
path="C:\\Users\lenovo\\Desktop\\img_align_celeba"
files=os.listdir(path)
global imgdir
imgdir=random.choice(files)

#Initiating Tkinter Window
root = Tk()    
root.geometry("650x670")

frame = Frame(root, width=600, height=400)
frame.pack()
frame.place(anchor='center', relx=0.5, rely=0.5)

img = ImageTk.PhotoImage(Image.open("C:\\Users\lenovo\\Desktop\\img_align_celeba\\" + imgdir))
img2 = ImageTk.PhotoImage(Image.open("C:\\Users\lenovo\\Desktop\\arousal_space.png"))

label = Label(frame, image = img, borderwidth=3, relief="groove")
label.pack()
label2 = Label(frame, image = img2, borderwidth=3, relief="groove")
label2.pack()

#Creating Default Row
global dataset
new_row = {'image_id':imgdir, 'Excited':-1, 'Delighted':-1, 'Happy':-1, 'Content':-1, 'Relaxed':-1, 'Calm':-1, 'Tired':-1, 'Bored':-1, 'Depressed':-1, 'Frustrated':-1, 'Angry':-1, 'Tense':-1}
dataset = dataset.append(new_row, ignore_index=True)
print(dataset.head())

#Write into csv file at path
student_header = ['image_id', 'Excited', 'Delighted', 'Happy', 'Content', 'Relaxed', 'Calm', 'Tired', 'Bored', 'Depressed', 'Frustrated', 'Angry', 'Tense']
with open('C:\\Users\lenovo\\Desktop\\arousal_csv.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerow(student_header)

#Change attribute values depending on button click
def chooseQuadrantOne():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Excited']=1
    dataset.at[dataImageID.index.item(),'Delighted']=1
    dataset.at[dataImageID.index.item(),'Happy']=1
    print(dataset.head())
    
    btnQuadrantOne.configure(bg="blue", fg="white")
    
def chooseQuadrantTwo():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Tense']=1
    dataset.at[dataImageID.index.item(),'Angry']=1
    dataset.at[dataImageID.index.item(),'Frustrated']=1
    print(dataset.head())
    
    btnQuadrantTwo.configure(bg="blue", fg="white")
    
def chooseQuadrantThree():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Depressed']=1
    dataset.at[dataImageID.index.item(),'Bored']=1
    dataset.at[dataImageID.index.item(),'Tired']=1
    print(dataset.head())
    
    btnQuadrantThree.configure(bg="blue", fg="white")

def chooseQuadrantFour():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Content']=1
    dataset.at[dataImageID.index.item(),'Relaxed']=1
    dataset.at[dataImageID.index.item(),'Calm']=1
    print(dataset.head())
    
    btnQuadrantFour.configure(bg="blue", fg="white")
        
def chooseExcited():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Excited']=1
    print(dataset.head())
    
    btnExcited.configure(bg="blue", fg="white")
    
def chooseDelighted():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Delighted']=1
    print(dataset.head())
    
    btnDelighted.configure(bg="blue", fg="white")
    
def chooseHappy():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Happy']=1
    print(dataset.head())
    
    btnHappy.configure(bg="blue", fg="white")
    
def chooseTense():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Tense']=1
    print(dataset.head())
    
    btnTense.configure(bg="blue", fg="white")

def chooseAngry():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Angry']=1
    print(dataset.head())
    
    btnAngry.configure(bg="blue", fg="white")
    
def chooseFrustrated():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Frustrated']=1
    print(dataset.head())
    
    btnFrustrated.configure(bg="blue", fg="white")

def chooseContent():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Content']=1
    print(dataset.head())
    
    btnContent.configure(bg="blue", fg="white")
    
def chooseRelaxed():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Relaxed']=1
    print(dataset.head())
    
    btnRelaxed.configure(bg="blue", fg="white")
    
def chooseCalm():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Calm']=1
    print(dataset.head())
    
    btnCalm.configure(bg="blue", fg="white")

def chooseDepressed():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Depressed']=1
    print(dataset.head())
    
    btnDepressed.configure(bg="blue", fg="white")

def chooseBored():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Bored']=1
    print(dataset.head())
    
    btnBored.configure(bg="blue", fg="white")
    
def chooseTired():
    global dataset
    dataImageID = dataset.query("image_id == @imgdir")
    dataset.at[dataImageID.index.item(),'Tired']=1
    print(dataset.head())
    
    btnTired.configure(bg="blue", fg = "white")
    
def saveToCSVFile():
    global dataset
    dataset.to_csv(r'C:\\Users\lenovo\\Desktop\\arousal_csv.csv')

    #Display next random image and reset button colors
def displayNewImage():
    path="C:\\Users\lenovo\\Desktop\\img_align_celeba"
    files=os.listdir(path)
    global imgdir
    imgdir=random.choice(files)
    updated_picture = ImageTk.PhotoImage(Image.open("C:\\Users\lenovo\\Desktop\\img_align_celeba\\" + imgdir))
    label.configure(image = updated_picture)
    label.photo_ref = updated_picture
    dataImageID = dataArousal.query("image_id == @imgdir")
    
    global dataset
    new_row = {'image_id':imgdir, 'Excited':-1, 'Delighted':-1, 'Happy':-1, 'Content':-1, 'Relaxed':-1, 'Calm':-1, 'Tired':-1, 'Bored':-1, 'Depressed':-1, 'Frustrated':-1, 'Angry':-1, 'Tense':-1}
    dataset = dataset.append(new_row, ignore_index=True)
    print(dataset.head())
    
    btnQuadrantOne.configure(bg= "#f0f0f0", fg = "black")
    btnQuadrantTwo.configure(bg= "#f0f0f0", fg = "black")
    btnQuadrantThree.configure(bg= "#f0f0f0", fg = "black")
    btnQuadrantFour.configure(bg= "#f0f0f0", fg = "black")
    btnExcited.configure(bg= "#f0f0f0", fg = "black")
    btnDelighted.configure(bg= "#f0f0f0", fg = "black")
    btnHappy.configure(bg= "#f0f0f0", fg = "black")
    btnTense.configure(bg= "#f0f0f0", fg = "black")
    btnAngry.configure(bg= "#f0f0f0", fg = "black")
    btnFrustrated.configure(bg= "#f0f0f0", fg = "black")
    btnContent.configure(bg= "#f0f0f0", fg = "black")
    btnRelaxed.configure(bg= "#f0f0f0", fg = "black")
    btnCalm.configure(bg= "#f0f0f0", fg = "black")
    btnDepressed.configure(bg= "#f0f0f0", fg = "black")
    btnBored.configure(bg= "#f0f0f0", fg = "black")
    btnTired.configure(bg= "#f0f0f0", fg = "black")

#Initialize buttons
btnQuadrantOne = Button(root, text = 'CHOOSE I', bd = '5', 
                          command=chooseQuadrantOne)

btnQuadrantTwo = Button(root, text = 'CHOOSE II', bd = '5',
                          command=chooseQuadrantTwo)

btnQuadrantThree = Button(root, text = 'CHOOSE III', bd = '5',
                          command=chooseQuadrantThree)

btnQuadrantFour = Button(root, text = 'CHOOSE IV', bd = '5', 
                          command=chooseQuadrantFour)

btnExcited = Button(root, text = 'EXCITED', bd = '5', 
                          command=chooseExcited)

btnDelighted = Button(root, text = 'DELIGHTED', bd = '5', 
                          command=chooseDelighted)

btnHappy = Button(root, text = 'HAPPY', bd = '5', 
                          command=chooseHappy)

btnTense = Button(root, text = 'TENSE', bd = '5', 
                          command=chooseTense)

btnAngry = Button(root, text = 'ANGRY', bd = '5', 
                          command=chooseAngry)

btnFrustrated = Button(root, text = 'FRUSTRATED', bd = '5', 
                          command=chooseFrustrated)

btnContent = Button(root, text = 'CONTENT', bd = '5', 
                          command=chooseContent)

btnRelaxed = Button(root, text = 'RELAXED', bd = '5', 
                          command=chooseRelaxed)

btnCalm = Button(root, text = 'CALM', bd = '5', 
                          command=chooseCalm)

btnDepressed = Button(root, text = 'DEPRESSED', bd = '5', 
                          command=chooseDepressed)

btnBored = Button(root, text = 'BORED', bd = '5', 
                          command=chooseBored)

btnTired = Button(root, text = 'TIRED', bd = '5', 
                          command=chooseTired)

btnNext = Button(root, text = 'NEXT', bd = '5', 
                          command=displayNewImage)

btnSaveToCSVFile = Button(root, text = 'SAVE', bd = '5', 
                          command=saveToCSVFile)

#Display buttons
btnQuadrantOne.place(x=525, y=260, bordermode=OUTSIDE)
btnQuadrantTwo.place(x=55, y=260, bordermode=OUTSIDE)
btnQuadrantThree.place(x=55, y=500, bordermode=OUTSIDE)
btnQuadrantFour.place(x=525, y=500, bordermode=OUTSIDE)
btnExcited.place(x=350, y=310, width=80, bordermode=OUTSIDE)
btnDelighted.place(x=380, y=355, width=80, bordermode=OUTSIDE)
btnHappy.place(x=400, y=400, width=80, bordermode=OUTSIDE)
btnTense.place(x=225, y=310, width=80, bordermode=OUTSIDE) 
btnAngry.place(x=190, y=355, width=80, bordermode=OUTSIDE)
btnFrustrated.place(x=170, y=400, width=80, bordermode=OUTSIDE)
btnContent.place(x=400, y=470, width=80, bordermode=OUTSIDE)
btnRelaxed.place(x=380, y=520, width=80, bordermode=OUTSIDE)
btnCalm.place(x=350, y=563, width=80, bordermode=OUTSIDE)
btnDepressed.place(x=170, y=470, width=80, bordermode=OUTSIDE)
btnBored.place(x=190, y=520, width=80, bordermode=OUTSIDE)
btnTired.place(x=225, y=565, width=80, bordermode=OUTSIDE)
btnNext.place(x=415, y=1, width=230, height=230, bordermode=OUTSIDE)
btnSaveToCSVFile.place(x=1, y=1, width=230, height=230, bordermode=OUTSIDE)

#Set default button colors
btnQuadrantOne.configure(bg= "#f0f0f0")
btnQuadrantTwo.configure(bg= "#f0f0f0")
btnQuadrantThree.configure(bg= "#f0f0f0")
btnQuadrantFour.configure(bg= "#f0f0f0")
btnExcited.configure(bg= "#f0f0f0")
btnDelighted.configure(bg= "#f0f0f0")
btnHappy.configure(bg= "#f0f0f0")
btnTense.configure(bg= "#f0f0f0")
btnAngry.configure(bg= "#f0f0f0")
btnFrustrated.configure(bg= "#f0f0f0")
btnContent.configure(bg= "#f0f0f0")
btnRelaxed.configure(bg= "#f0f0f0")
btnCalm.configure(bg= "#f0f0f0")
btnDepressed.configure(bg= "#f0f0f0")
btnBored.configure(bg= "#f0f0f0")
btnTired.configure(bg= "#f0f0f0")
 

root.mainloop()
```
